__Import Necessary Libraries__

```python
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt 
```

__Loading of Diabetes dataset__

```python
from sklearn import datasets
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

__Exploration of Dataset__

_diabetes_X consists of 10 features: age, sex, body mass index, average blood pressure, and six blood serum requirements._

_diabetes_Y is a quantitative measure of disease progression one year after baseline._

```python
plt.scatter(diabetes_X[:,0], diabetes_y)
plt.xlabel("Age")
plt.ylabel("Disease Progression")
plt.title("Age vs. Disease Progression")
plt.show()
```

```python
X_BMI = diabetes_X[:,np.newaxis,2]
```

__Splitting the data into training and testing__

```python
X_train, X_test, y_train, y_test = train_test_split(X_BMI, diabetes_y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()  
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print('Mean Absolute error: %.2f' % metrics.mean_absolute_error(y_test, y_pred))
print('Mean squared error: %.2f' % metrics.mean_squared_error(y_test, y_pred))
print('R-Squared value: %.2f' % r2_score(y_test,y_pred))
```

```python
plt.scatter(X_train,y_train, color='green')
plt.scatter(X_test, y_test,  color='red')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
```

_On the basis of MAE, MSE and R-Squared value, it does not seem to be a good fit._
